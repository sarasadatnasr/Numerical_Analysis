## Problem 1
Consider an arbitrary hat function. Plot T(x) and main hat function. 
You should solve (calculate T(x)) it by python.


```python
def h(x):
    if abs(x) < 1:
        return 1 - abs(x)
    else:
        return 0
```

```python
def T(x):
    return (h(x-1) + h(x) + h(x+1))/3
```

```python
xx = np.linspace(-5, 5, num=1000)
y_h = [h(x) for x in xx]
y_T = [T(x) for x in xx]
```

```python
fig, axs = plt.subplots(figsize=(10, 5))
axs.plot(xx, y_h, label='h(x)')
axs.plot(xx, y_T, label='T(x)')
axs.legend()
plt.tight_layout()
plt.show()
```

## Conclusion for this problem
In this code, we first define the hat function h(x) as a piecewise linear function that takes the value of $1 - |x|$ for $|x| < 1$ and 0 otherwise.

We then define the tent function T(x) as the average of three shifted and scaled copies of the hat function.
