Brent oil price in period of time - log plot

```python
import pandas as pd
df = pd.read_csv('BrentOilPrices.csv')
print(df.to_string()) 
```

```python
import matplotlib.pyplot as plt
data = df['Price']
fig, ax = plt.subplots()
ax.plot(data)
ax.set_yscale('log')
plt.show()
```
