## Problem 1
Plot each one of these functions with their error in an appropriate interval:

```python
import matplotlib.pyplot as plt
import numpy as np
import math
```

```python
def taylorSin(n,x):
    result = 0
    for i in range(n):
        p = (2*i)+1
        if i%2 == 0:
            result+= x**(p)/math.factorial(p)
        else:
            result+= -1*x**(p)/math.factorial(p)   
    return result
```

```python
def taylorCos(n,x):
    result = 0
    for i in range(n):
        p = (2*i)
        if i%2 == 0:
            result += x**(p)/math.factorial(p)
        else:
            result += -1*x**(p)/math.factorial(p)   
    return result
```

### Part a)
$sin(x) , x $

```python
x = np.arange(-1*math.pi, math.pi,0.05)
y = np.sin(x)
taylor = taylorSin(1,x)
error = abs(y-taylor)
plt.plot(x, y, color='g', label='sin')
plt.plot(x, taylor ,color='b', label='taylor series')
plt.plot(x, error ,color='r', label='error')
plt.xlabel('x')
plt.ylabel('y')
plt.title("Sine and Taylor functions")
plt.legend()
plt.show()
```

## Problem 2
Provide a polynomial based on Taylor series to interpolate function f(x) = cos(x). Your suggested
polynomial should has error lower than 10âˆ’7

```python
import math

def cosineEstimator(x):
    x %= 2 * math.pi
    c = 2
    n = 0
    f = 2
    for i in range(30):
        if i % 2 == 0:
            n += x**c / f
        else:
            n -= x**c / f
        c += 2
        f *= c * (c - 1)
    return 1 - n
print('The actual value in x = 5 is equal to',math.cos(5))
print('The estimated value is equal to',cosineEstimator(5))
print('The error obtained from this polynomial is equal to:',cosineEstimator(5)-math.cos(5))    
```
